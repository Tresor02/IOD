````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b 

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X, Y);
yfit = [b + m * xi for xi in X]
plt.plot(X, yfit);
plt.show()
````

# 13. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem

Expected output:

(172.52, 168.26329573448135, 176.77670426551867)

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, se

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))
````
